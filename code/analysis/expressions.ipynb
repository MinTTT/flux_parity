```python
import numpy as np 
import pandas as pd 
import bokeh.io 
import bokeh.plotting 
import bokeh.layouts
import growth.model
bokeh.io.output_notebook()
```

```python
gamma_max = 9.65
nu_max = 8 
Kd = 0.025

phiR_range = np.linspace(0.001, 0.999999, 1000)
Gamma = gamma_max * phiR_range
Nu = nu_max * (1 - phiR_range)


cAA_ss = ((Nu - Gamma) + np.sqrt((Gamma - Nu)**2 + 4 * Gamma * Nu * Kd)) / (2 * Gamma)
mu_ss = Gamma / (1 + (Kd * 2 * Gamma) / ((Nu - Gamma) + np.sqrt((Gamma - Nu)**2 + 4 * Gamma * Nu * Kd)))

mu_test = (Gamma + Nu - np.sqrt((Gamma + Nu)**2 - 4 * (1 - Kd) * Gamma * Nu))/(2 * (1 - Kd))
```

```python
p1 = bokeh.plotting.figure()
p2 = bokeh.plotting.figure(y_axis_type='log')
p1.line(x=phiR_range, y=mu_ss)
p1.line(x=phiR_range, y=mu_test, color='red')
p2.line(x=phiR_range, y=cAA_ss)
p1.xaxis.axis_label = r'\[\phi_R\]'
p2.xaxis.axis_label = r"$$\phi_R$"
p1.yaxis.axis_label = r'$$\mu$$'
p2.yaxis.axis_label = r'$cAA$'
row = bokeh.layouts.row(p1, p2)
bokeh.io.show(row)
```
